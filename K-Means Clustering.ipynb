```python
import numpy as np
import math
import random
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from matplotlib  import cm
```

```python
def euclidean_distance(x1,x2):
    result = np.sqrt(np.sum(((x1-x2)**2)))
    return result
        
class KMeans:
    def __init__(self,clusters=5,iterations=100):
        self.clusters_size = clusters
        self.iterations = iterations
        self.clusters = None
        self.centroid = None
        
    def new_centroid(self):
        idx = 0
        for i in self.clusters:
            temp_centroid = np.mean(i,axis=0)
            cen = 0
            distance = np.inf
            for j in i:
                temp = euclidean_distance(temp_centroid,j)
                if temp < distance:
                    distance = temp
                    cen = j
            self.centroid[idx] = cen
            idx +=1
        
    def classification(self,X):
        n_samples,n_features = X.shape
        self.centroid = X[np.random.choice(X.shape[0],size = self.clusters_size,replace=False)]
        for x in range(self.iterations):
            if x != 0:
                self.new_centroid()
            self.clusters = []
            for k in range(self.clusters_size):
                ele = []
                self.clusters.append(ele)
            for i in range(n_samples):
                distance = np.inf
                cen = 0
                idx = 0
                for j in self.centroid:
                    temp = euclidean_distance(X[i],j)
                    if temp < distance:
                        distance = temp
                        cen = idx
                    idx += 1
                self.clusters[cen].append(X[i])
                
    def get_centroids(self):
        return self.centroid
    
    def get_clusters(self):
        return self.clusters
        
    
```

```python
# n_samples = input("Enter the sample size: ")
# n_features = input("Enter the num of features: ")
# X = np.random.rand(int(n_samples),int(n_features))
Y= []
with open("points.txt") as p:
    for line in p:
        Y.append([int(x)/10 for x in line.split()])
    
X = np.array(Y)
clustering = KMeans()
clustering.classification(X)
centroids = clustering.get_centroids()
clusters = clustering.get_clusters()
print(centroids)
print("----------------------------------------------")
clusters
```
