```python
import numpy as np
import math
from sklearn.model_selection import train_test_split
from sklearn import datasets
from collections import Counter
```

```python
class NaiveBayes:
    def __init__(self):
        self.classes = None
        self.likelihood = None
        self.likelihood_appearance = None
        self.class_appearance = {}
        self.frequency_table = []
    
    def to_frequency_table(self,X,y):
        n_samples, n_features = X.shape
        self.likelihood = np.unique(y)
        self.classes = np.unique(X,axis=0)
        for val in self.classes:
            new_val = tuple(val)
            indices = [i for i, x in enumerate(X) if (x==val).all()]
            self.class_appearance[new_val] = indices
        for i in self.classes:
            class_ele = []
#             class_ele.append(i)
            class_info = {}
            for k in self.likelihood:
                class_info[k] = 0
            it = tuple(i)
            for j in self.class_appearance[it]:
                class_info[y[j]] += 1
            class_ele.append(class_info)
            self.frequency_table.append(class_ele)
        return self.frequency_table
        
    
    def train(self,X,y):
        self.to_frequency_table(X,y)
        self.likelihood_appearance = Counter(y)
        n_samples,n_features = X.shape
        for i in self.frequency_table:
            event_num = 0
            for key,val in self.likelihood_appearance.items():
                event_num += i[0][key]
                i[0][key] /= val
            i.append(event_num / n_samples)
        for key,val in self.likelihood_appearance.items():
            self.likelihood_appearance[key] = val / n_samples
    
    def predict(self,X):
        ret = []
        for i in X:
            idx = 0
            for k in self.classes:
                if ((k==i).all()):
                    break
                idx +=1
            key_max = 0
            val_max = 0
#             print(self.frequency_table[idx])
            for key,val in self.frequency_table[idx][0].items():
                temp = self.frequency_table[idx][0][key]*self.likelihood_appearance[key]/self.frequency_table[idx][1]
                if temp > val_max:
                    key_max = key
                    val_max = temp
            ret.append(key_max)
        return ret
            
            
        
```

```python
X_train = []
for i in range(10000):
    ele = np.random.choice(5,4,replace=True)
    X_train.append(ele)

X_train = np.array(X_train)
y_train = np.random.choice(2,10000,replace = True)
```
